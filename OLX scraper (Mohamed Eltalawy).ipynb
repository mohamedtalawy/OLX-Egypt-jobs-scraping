```python
# importing packages
import pandas as pd
import re

from bs4 import Tag, NavigableString, BeautifulSoup as bs
from datetime import date, timedelta, datetime
from IPython.core.display import clear_output
from random import randint
from requests import get
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
from time import time
import datetime
start_time = time()
from itertools import zip_longest
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

from warnings import warn
```

### 2- Identifing the used lists:

```python
links=[]
titles=[]
locations=[]
titles=[]
datestimes=[]
levels=[]
edulevels=[]
emptypes=[]
salaries=[]
industries=[]
descriptions=[]
```

### 3- Scraping the pages with filters in OLX.com:

```python
# Tell Selenium to get the URL we're interested in.
driver = webdriver.Chrome()
driver.maximize_window()
action = ActionChains(driver)

# Scraping multiple pages
for x in range(23):
    result= driver.get(f'https://www.olx.com.eg/en/jobs/engineering/?search%5Bfilter_enum_type%5D%5B0%5D=1&search%5Bfilter_enum_field%5D%5B0%5D=2&search%5Bfilter_enum_field%5D%5B1%5D=3&search%5Bfilter_enum_field%5D%5B2%5D=4&search%5Bfilter_enum_field%5D%5B3%5D=5&search%5Bfilter_enum_field%5D%5B4%5D=6&search%5Bfilter_enum_field%5D%5B5%5D=7&search%5Bfilter_enum_field%5D%5B6%5D=8&search%5Bfilter_enum_field%5D%5B7%5D=9&search%5Bfilter_enum_field%5D%5B8%5D=10&view=list&page={x+1}')
    timeout = 2
    try:
        element_present = EC.presence_of_element_located((By.ID, 'element_id'))
        WebDriverWait(driver, timeout).until(element_present)
    except TimeoutException:
        print(f"Page {x+1} has been scrolled down")
    source_data = driver.page_source
    soup = bs(source_data)
    link= soup.find_all("div" , {"class":"ads__item__info"})
    for i in range(len(link)):
        links.append(link[i].find("a").attrs['href'])
print("There are available",len(links),"jobs")
```

```python
for i in links:
    result= driver.get(i)
    source_data = driver.page_source
    
    title= driver.find_element_by_xpath('//div[@class="clr offerheadinner pding15 pdingright20"]/h1')
    titles.append(title.text.strip())
    
    location= driver.find_element_by_xpath('//span[@class="show-map-link link gray cpointer"]/strong')
    locations.append(location.text.strip())
    
    edulevel= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[2]/td[1]/table/tbody/tr/td/strong')
    edulevels.append(edulevel.text.strip())
                                           
    datetime= driver.find_element_by_xpath('//small[@class="small c62 lheight22 inlblk"]/span[@class ="pdingleft10 brlefte5"]')
    datestimes.append(datetime.text.split("رقم",1)[0].split("في",1)[1].strip())
                                           
    level= driver.find_element_by_xpath('//td[@class="value"]/strong')
    levels.append(level.text.strip())
                                           
    emptype= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr/td[2]/table/tbody/tr/td/strong')
    emptypes.append(emptype.text.strip())

    salary= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[2]/td[2]/table/tbody/tr/th').text.strip()
    if salary  == "الراتب" :
        salary1= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[2]/td[2]/table/tbody/tr/td/strong')
        salaries.append(salary1.text.replace(",","").strip())
    else:
        salaries.append("Confidential") 
        
    industry= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[2]/td[2]/table/tbody/tr/th').text.strip()
    if industry  == "نوع المُعلن" :
        industry1= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[2]/td[2]/table/tbody/tr/td/strong')
        industries.append(industry1.text.strip())
    else:
        industry1= driver.find_element_by_xpath('//div[@class="clr descriptioncontent marginbott20"]/table/tbody/tr[3]/td/table/tbody/tr/td/strong')
        industries.append(industry1.text.strip())  
    description= driver.find_element_by_xpath('//p[@class="pding10 lheight20 large"]')
    descriptions.append(description.text.strip())

print("Titles list contains:",len(titles)) 
print("Seniority levels list contains:",len(levels)) 

print("Job Titles Sample:",titles[:5])
print("Job location Sample:",locations[:5])
print("Date and Time Sample:",datestimes[:5])
print("Seniority levels Sample:",levels[:5])
print("Levels of experience Sample:",edulevels[:5])
print("Employment type Sample:",emptypes[:5])
print("Salaries Sample:",salaries[:5])
print("Sector and Industries Sample:",industries[:5])
```

```python
# to check if we have all information
print(len(titles))
print(len(locations))
print(len(datestimes))
print(len(salaries))
print(len(emptypes))
print(len(descriptions))
print(len(levels))
print(len(industries))
print(len(edulevels))
```

```python
# creating a dataframe
job_data_olx = pd.DataFrame({'Job Title': titles,
'Location': locations,
'Date and Time': datestimes,
'Seniority Level': levels,
'Education levels': edulevels,
'Type': emptypes,
'Salary': salaries,
'Industry': industries,
'Description': descriptions,
})

# cleaning description column
job_data_olx['Description'] = job_data_olx['Description'].str.replace('\n',' ')
print(job_data_olx.info())
job_data_olx.head()
```
